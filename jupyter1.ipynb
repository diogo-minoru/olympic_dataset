```python
import pandas as pd
import numpy as np

olympic_df = pd.read_csv(r"./athletes/bios.csv", usecols = ["Roles", "Sex", "Used name", "Born", "Died", "NOC", "athlete_id", "Measurements", "Affiliations"])
olympic_df.head()
```

```python
olympic_df.shape
```

#### Percental de valores que estão nulos em cada coluna

```python
olympic_df.isna().apply(lambda x: (x.sum() / len(olympic_df)) * 100)
```

#### Remover caracteres especiais da coluna "Used Name"

```python
#olympic_df[olympic_df["Used name"].str.contains("•")]
olympic_df["Used name"] = olympic_df["Used name"].str.replace("•", " ")
olympic_df.head()
```

#### Adicionando coluna de datas para a data de nascimento e data de falecimento

```python
olympic_df["BornDate"] = pd.to_datetime(olympic_df["Born"].str.split(" ").str[0:3].str.join("/"), format = '%d/%B/%Y', errors = "coerce")
olympic_df["BornYear"] = olympic_df["BornDate"].dt.year.astype("Int64")
olympic_df["BornDate"] = olympic_df["BornDate"].dt.strftime('%d/%m/%Y')
olympic_df.head()
```

```python
olympic_df["DiedDate"] = pd.to_datetime(olympic_df["Died"].str.split(" ").str[0:3].str.join("/"), format = '%d/%B/%Y', errors = "coerce")
olympic_df["DiedYear"] = olympic_df["DiedDate"].dt.year.astype("Int64")
olympic_df["DiedDate"] = olympic_df["DiedDate"].dt.strftime('%d/%m/%Y')
olympic_df.head()
```

#### Separar coluna em altura e peso

```python
olympic_df[["height_cm", "weight_kg"]] = olympic_df["Measurements"].str.split("/", expand = True)
```

```python
olympic_df[
    olympic_df["Measurements"].notna() & 
    (olympic_df["height_cm"].str.contains("kg", na = False) | 
     olympic_df["weight_kg"].str.contains("cm", na = False))]
```

#### Tratar casos em que a coluna de medidas possui apenas a altura ou apenas o peso
Quando há apenas um registro  (peso ou altura) este será armazenado na coluna "height_cm" pois não há o delimitador "/"
Solução:
Quando a coluna de altura possuir a string "kg", este valor deverá ser atribuído à coluna "weight_kg", após isso, deverá ser substituído por nulo os valores da coluna "height_cm" com "kg"
Quando a coluna 

```python
olympic_df["weight_kg"] = np.where(olympic_df["height_cm"].str.contains("kg", na = False), olympic_df["height_cm"], olympic_df["weight_kg"])
```

```python
olympic_df["height_cm"] = np.where(olympic_df["height_cm"].str.contains("kg", na = False), np.nan, olympic_df["height_cm"])
```

```python
# Fazendo verificação se a alteração deu certo

olympic_df[
    olympic_df["Measurements"].notna() & 
    (olympic_df["height_cm"].str.contains("kg", na = False) | 
     olympic_df["weight_kg"].str.contains("cm", na = False))]
```

```python
# Não há registro de informações de altura na coluna "weight_kg"
olympic_df[olympic_df["weight_kg"].str.contains("cm", na = False)]
```

```python
# Não há registro de informações de peso na coluna "height_cm"
olympic_df[olympic_df["height_cm"].str.contains("kg", na = False)]
```
