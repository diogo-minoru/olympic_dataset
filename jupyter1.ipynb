```python
import pandas as pd

olympic_df = pd.read_csv(r"./athletes/bios.csv", usecols = ["Roles", "Sex", "Used name", "Born", "Died", "NOC", "athlete_id", "Measurements", "Affiliations"])
olympic_df.head()
```

```python
olympic_df.shape
```

#### Percental de valores que estão nulos em cada coluna

```python
olympic_df.isna().apply(lambda x: (x.sum() / len(olympic_df)) * 100)
```

#### Remover caracteres especiais da coluna "Used Name"

```python
#olympic_df[olympic_df["Used name"].str.contains("•")]
olympic_df["Used name"] = olympic_df["Used name"].str.replace("•", " ")
olympic_df.head()
```

#### Adicionando coluna de datas para a data de nascimento e data de falecimento

```python
olympic_df["BornDate"] = pd.to_datetime(olympic_df["Born"].str.split(" ").str[0:3].str.join("/"), format = '%d/%B/%Y', errors = "coerce")
olympic_df["BornYear"] = olympic_df["BornDate"].dt.year.astype("Int64")
olympic_df["BornDate"] = olympic_df["BornDate"].dt.strftime('%d/%m/%Y')
olympic_df.head()
```

```python
olympic_df["DiedDate"] = pd.to_datetime(olympic_df["Died"].str.split(" ").str[0:3].str.join("/"), format = '%d/%B/%Y', errors = "coerce")
olympic_df["DiedYear"] = olympic_df["DiedDate"].dt.year.astype("Int64")
olympic_df["DiedDate"] = olympic_df["DiedDate"].dt.strftime('%d/%m/%Y')
olympic_df.head()
```
